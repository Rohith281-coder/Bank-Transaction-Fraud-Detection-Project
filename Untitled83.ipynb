```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"C:\Users\rohit\Downloads\archive\bank_transactions_data_2.csv")
print(df)
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['PreviousTransactionDate'] = pd.to_datetime(df['PreviousTransactionDate'])

print("Dataset Shape:", df.shape)
print(df.head(2))
```

```python
eda_summary = {
    "Total Transactions": len(df),
    "Unique Accounts": df['AccountID'].nunique(),
    "Unique Merchants": df['MerchantID'].nunique(),
    "Avg Transaction Amount": df['TransactionAmount'].mean(),
    "Max Transaction Amount": df['TransactionAmount'].max(),
    "Min Transaction Amount": df['TransactionAmount'].min(),
}
print("\nEDA Summary:", eda_summary)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['TransactionAmount'], bins=50, color="skyblue", edgecolor="black")
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['Channel'].value_counts().plot(kind="bar", color="orange", edgecolor="black")
plt.title("Transactions by Channel")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby(df['TransactionDate'].dt.date).size().plot(color="green")
plt.title("Transactions Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

```python
features = ["TransactionAmount", "TransactionDuration", "LoginAttempts", "AccountBalance"]
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
iso = IsolationForest(contamination=0.03, random_state=42)
df['Anomaly'] = iso.fit_predict(X_scaled)
df['Anomaly'] = df['Anomaly'].map({1:0, -1:1})  
print("\nTotal anomalies detected:", df['Anomaly'].sum())
print("Anomaly rate: {:.2f}%".format(df['Anomaly'].mean()*100))
```

```python
plt.figure(figsize=(6,4))
channel_anomalies = df.groupby(["Channel","Anomaly"]).size().unstack(fill_value=0)
channel_anomalies.plot(kind="bar", stacked=True, color=["blue","red"], edgecolor="black")
plt.title("Anomalies by Channel")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
location_counts = df[df['Anomaly']==1]['Location'].value_counts().head(10)
location_counts.plot(kind="barh", color="crimson", edgecolor="black")
plt.title("Top 10 Locations with Fraudulent Transactions")
plt.xlabel("Fraud Count")
plt.gca().invert_yaxis()  # Highest on top
plt.show()
```

```python
plt.figure(figsize=(8,6))
normal = df[df['Anomaly']==0]
fraud = df[df['Anomaly']==1]
plt.scatter(normal['TransactionAmount'], normal['AccountBalance'], c="blue", alpha=0.5, label="Normal")
plt.scatter(fraud['TransactionAmount'], fraud['AccountBalance'], c="red", alpha=0.7, label="Fraud")
plt.title("Transaction Amount vs Account Balance (Fraud Highlighted)")
plt.xlabel("Transaction Amount")
plt.ylabel("Account Balance")
plt.legend()
plt.show()
```

```python
df.to_csv("bank_transactions_with_anomalies.csv", index=False)
print("\nProcessed file saved: bank_transactions_with_anomalies.csv")
```
